```python
import folium
import geopandas as gpd
import pandas as pd
import GeoData
```

```python
in_lebanon = gpd.read_file("GeoData/lebanon.shp")

gov = in_lebanon[["geometry", "name_en"]]
x = gpd.read_file("GeoData/akkar_fixed.shp")
x["name_en"]= "Akkar"
x = x[["geometry", "name_en"]]
overall = pd.concat( [gov,x], ignore_index=True)
```

```python
map = folium.Map(location=[33.8938, 35.5018], tiles="OpenStreetMap", zoom_start=8)
```

```python
# folium.GeoJson(data=overall["geometry"],
#                highlight_function= lambda feat: {'fillColor': 'blue'}
#               ).add_to(map)
```

```python
geo_json = folium.GeoJson(overall,  
                          highlight_function= lambda feat: {'fillColor': 'blue'}, 
                          popup=folium.GeoJsonPopup(fields=['name_en'],labels=False ))

geo_json.add_to(map)
```

```python
data = [['beirut', "html/beirut.html",33.8869, 35.5131], 
        ['baalbek', "html/baalbek.html",34.265856, 36.34981], 
        ['north', "html/north.html", 34.438063,35.830823], 
        ['beqaa', "html/beqaa.html", 33.846266,35.901949], 
        ['akkar', "html/akkar.html",34.532876,36.132813], 
        ['south', "html/south.html",33.272148,35.203278], 
        ['mount', "html/mount.html",33.810086,35.597314], 
        ['nabatieh', "html/nabatieh.html",33.377169,35.483829]]

df = pd.DataFrame(data, columns=['Name', 'html_file','Latitude','Longitude'])
```

```python
df
```

```python
for i in range(0,len(df)):
    html="""
    <iframe src=\"""" + df['html_file'][i] + """\" width="850" height="400"  frameborder="0">    
    """
    
    popup = folium.Popup(folium.Html(html, script=True))
    folium.Marker([df['Latitude'].iloc[i],df['Longitude'].iloc[i]],
                  popup=popup,icon=folium.Icon(color='purple', icon='school', prefix='fa')).add_to(map)
    
#m.save("index.html")
#view raw
```
